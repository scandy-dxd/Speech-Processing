# **CNN**

```python
import librosa
import random
import numpy as np
import glob
features=[]
labels=[]
emotions=['OAF_Fear',
 'OAF_Pleasant_surprise',
 'OAF_Sad',
 'OAF_angry',
 'OAF_disgust',
 'OAF_happy',
 'OAF_neutral',
 'YAF_angry',
 'YAF_disgust',
 'YAF_fear',
 'YAF_happy',
 'YAF_neutral',
 'YAF_pleasant_surprised',
 'YAF_sad']


for spk in emotions:
  voice=glob.glob(f'/content/drive/MyDrive/Emotion_Recog/TESS Toronto emotional speech set data/{spk}/**/*.wav',recursive=True)

  for audio in voice:
    y,sr=librosa.load(audio)
    mels=librosa.feature.melspectrogram(y=y,n_mels=128,sr=sr,)
    mels=mels.T
    features.append(mels)
    labels.append(spk)
max_length = max(f.shape[0] for f in features)
padded_features=[np.pad(f,((0,max_length-f.shape[0]),(0,0)),mode='constant')for f in features]








```

```python
import numpy as np
x=np.array(padded_features)
y=np.array(labels)

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y= label_encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split


X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
segments=X_train.shape[1]
features=X_train.shape[2]
```

```python
from tensorflow import keras

# Assuming X_train, X_test, y_train are defined

# Reshape input data
X_train = X_train.reshape(-1, segments, features)
X_test = X_test.reshape(-1, segments, features)

# Define the model
model = keras.Sequential()
model.add(keras.layers.Conv2D(32,(3,3),activation='relu',input_shape=(segments,features,1)))
model.add(keras.layers.MaxPooling2D(2,2))
# model.add(keras.layers.Conv2D(32,(3,3),activation='relu'))
# model.add(keras.layers.MaxPooling2D(2,2))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(64,activation='relu'))
model.add(keras.layers.Dropout(0.1))
model.add(keras.layers.Dense(14,activation='softmax'))



# Compile the model
optimiser = keras.optimizers.Adam(learning_rate=0.0001)
model.compile(optimizer=optimiser, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Print model summary
model.summary()


# Train the model
history = model.fit(X_train, y_train, epochs=60, batch_size=32,validation_data=(X_test,y_test),callbacks=[early_stopping])

test_loss,test_acc=model.evaluate(X_test,y_test)
print("Test Accuracy:",test_acc)
```
