Converting the sound to Numpy Array

```python
y,sr=librosa.load(os.listdir()[2], duration=2.86)
ps=librosa.feature.melspectrogram(y=y, sr=sr)
ps.shape
```

Visualising the data using librosa.display

```python
librosa.display.specshow(ps,y_axis='mel', x_axis='time')
```

To plot wave form , use plt.plot(y) , on librosa we cannot plot waveplot

```python
X= librosa.stft(y)
Xdb =librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14,5))
librosa.display.specshow(Xdb,sr=sr,x_axis='time',y_axis='hz')
```
